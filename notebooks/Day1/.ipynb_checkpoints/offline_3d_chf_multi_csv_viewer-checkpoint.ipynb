# Offline 3D CHF multi-CSV viewer — JupyterLab version

This notebook is a Python/JupyterLab conversion of the custom functions from the attached HTML viewer.

It preserves the main workflow:

1. Load one or more CSV files.
2. Add synthetic columns for file name and original CSV row.
3. Guess likely source, X, Y, Z, and colour columns.
4. Clean numeric values.
5. Optionally plot `ln(colour value)`.
6. Render an interactive Plotly 3D scatter plot grouped by source.

The original browser-only controls have been converted to notebook functions and optional `ipywidgets` controls.

````python
from __future__ import annotations

from dataclasses import dataclass
from pathlib import Path
import io
from typing import Iterable, Sequence

import numpy as np
import pandas as pd
import plotly.graph_objects as go

try:
    import ipywidgets as widgets
    from IPython.display import display, clear_output
except Exception:
    widgets = None
    display = None
    clear_output = None
````

## Constants and utility functions

These are Python equivalents of the HTML/JavaScript constants and helper functions such as `normalise`, `findHeader`, `toNumber`, and the colour scale selector.

````python
SYNTHETIC_FILE_COL = "__File name"
SYNTHETIC_ROW_COL = "__File row"

COLOR_SCALES = {
    "Cool": [[0.0, "rgb(0,255,255)"], [1.0, "rgb(255,0,255)"]],
    "Viridis": "Viridis",
    "Plasma": "Plasma",
    "Inferno": "Inferno",
    "Magma": "Magma",
    "Cividis": "Cividis",
    "Turbo": "Turbo",
    "Jet": "Jet",
    "Rainbow": "Rainbow",
    "Greys": "Greys",
}


def get_selected_color_scale(name: str = "Cool"):
    """Return a Plotly colour scale by name."""
    return COLOR_SCALES.get(name, COLOR_SCALES["Cool"])


def normalise(value) -> str:
    """Lowercase a label and collapse non-alphanumeric characters to spaces."""
    import re

    return re.sub(r"[^a-z0-9]+", " ", str(value).lower()).strip()


def find_header(headers: Sequence[str], required_terms: Sequence[str], avoid_terms: Sequence[str] = ()) -> str:
    """Find the first header containing all required terms and none of the avoid terms."""
    norms = [normalise(h) for h in headers]
    for header, norm in zip(headers, norms):
        has_all = all(term in norm for term in required_terms)
        avoids = any(term in norm for term in avoid_terms)
        if has_all and not avoids:
            return header
    return headers[0] if headers else SYNTHETIC_FILE_COL


def to_number(values) -> pd.Series:
    """Convert values to numbers after removing thousands separators and Unicode minus signs."""
    series = pd.Series(values, copy=False).astype(str)
    series = series.str.replace(",", "", regex=False).str.replace("−", "-", regex=False).str.strip()
    return pd.to_numeric(series, errors="coerce")


def guess_column_mapping(headers: Sequence[str]) -> dict[str, str]:
    """Replicate the HTML viewer's default column-guessing logic."""
    headers = list(headers)
    if not headers:
        return {}

    source_guess = find_header(headers, ["reference", "name"])
    if source_guess == headers[0]:
        source_guess = find_header(headers, ["reference", "id"])
    if source_guess == headers[0]:
        source_guess = find_header(headers, ["source"])
    if source_guess == headers[0] and SYNTHETIC_FILE_COL in headers:
        source_guess = SYNTHETIC_FILE_COL

    quality_guess = find_header(headers, ["outlet", "quality"])
    if quality_guess == headers[0]:
        quality_guess = find_header(headers, ["outlet", "qu"])

    return {
        "source": source_guess,
        "x": find_header(headers, ["mass", "flux"]),
        "y": find_header(headers, ["pressure"]),
        "z": quality_guess,
        "color": find_header(headers, ["chf"], ["result"]),
    }
````

## Viewer state and CSV loading

`CHFCSVViewer` replaces the browser's global state: `tableRows`, `headers`, and `loadedFiles`. It can load CSV files from paths or from an `ipywidgets.FileUpload` control.

````python
@dataclass
class LoadedFile:
    name: str
    rows: int


class CHFCSVViewer:
    """Notebook implementation of the HTML 3D CHF multi-CSV viewer."""

    def __init__(self):
        self.data = pd.DataFrame()
        self.headers: list[str] = []
        self.loaded_files: list[LoadedFile] = []
        self.has_rendered_plot = False

    def clear(self):
        """Clear all loaded data and reset the viewer state."""
        self.data = pd.DataFrame()
        self.headers = []
        self.loaded_files = []
        self.has_rendered_plot = False

    def _merge_headers(self, new_headers: Iterable[str]):
        combined = list(self.headers)
        for header in new_headers:
            if header not in combined:
                combined.append(header)
        for synthetic in (SYNTHETIC_FILE_COL, SYNTHETIC_ROW_COL):
            if synthetic not in combined:
                combined.append(synthetic)
        self.headers = combined

    @staticmethod
    def _read_csv_from_bytes(content: bytes) -> pd.DataFrame:
        return pd.read_csv(io.BytesIO(content), dtype=str, keep_default_na=False)

    @staticmethod
    def _read_csv_from_path(path: str | Path) -> pd.DataFrame:
        return pd.read_csv(path, dtype=str, keep_default_na=False)

    def _append_dataframe(self, df: pd.DataFrame, file_name: str) -> int:
        """Append a CSV DataFrame and add synthetic source metadata columns."""
        df = df.copy()
        df.columns = [str(c).strip() for c in df.columns]
        original_headers = list(df.columns)

        if df.empty:
            row_count = 0
        else:
            # Match the HTML rule: ignore rows where all fields collapse to '' or '-'.
            joined = df.astype(str).agg(" ".join, axis=1).str.strip()
            df = df.loc[~joined.isin(["", "-"])].copy()
            df[SYNTHETIC_FILE_COL] = file_name
            df[SYNTHETIC_ROW_COL] = (df.index.to_numpy() + 2).astype(str)  # +2 accounts for the header row.
            row_count = len(df)

        self._merge_headers([*original_headers, SYNTHETIC_FILE_COL, SYNTHETIC_ROW_COL])
        if row_count:
            self.data = pd.concat([self.data, df], ignore_index=True, sort=False)
            self.data = self.data.reindex(columns=self.headers)
        return row_count

    def load_files(self, file_paths: str | Path | Sequence[str | Path]) -> pd.DataFrame:
        """Load one or more CSV files by filesystem path."""
        if isinstance(file_paths, (str, Path)):
            file_paths = [file_paths]

        for file_path in file_paths:
            path = Path(file_path)
            try:
                df = self._read_csv_from_path(path)
                rows = self._append_dataframe(df, path.name)
                self.loaded_files.append(LoadedFile(path.name, rows))
            except Exception as exc:
                self.loaded_files.append(LoadedFile(f"{path.name} (failed to load: {exc})", 0))
        return self.data

    def load_uploaded(self, upload_value) -> pd.DataFrame:
        """Load files from an ipywidgets.FileUpload `.value` payload."""
        if isinstance(upload_value, dict):
            items = upload_value.values()
        else:
            items = upload_value or []

        for item in items:
            name = item.get("name", "uploaded.csv")
            content = item.get("content", b"")
            if isinstance(content, memoryview):
                content = content.tobytes()
            try:
                df = self._read_csv_from_bytes(bytes(content))
                rows = self._append_dataframe(df, name)
                self.loaded_files.append(LoadedFile(name, rows))
            except Exception as exc:
                self.loaded_files.append(LoadedFile(f"{name} (failed to load: {exc})", 0))
        return self.data

    def default_mapping(self) -> dict[str, str]:
        return guess_column_mapping(self.headers)

    def file_summary(self) -> pd.DataFrame:
        """Return a summary of loaded files and row counts."""
        return pd.DataFrame([file.__dict__ for file in self.loaded_files])
````

## Data cleaning and figure creation

These functions convert the HTML viewer's `renderPlot` logic into reusable Python functions. The Plotly figure is grouped by source, shares a common colour range across traces, and keeps the original file name and CSV row in hover text.

````python
def prepare_plot_data(
    data: pd.DataFrame,
    *,
    source_col: str,
    x_col: str,
    y_col: str,
    z_col: str,
    color_col: str,
    use_log_color: bool = False,
) -> tuple[pd.DataFrame, int]:
    """Clean selected columns and return plottable rows plus the log-exclusion count."""
    required = [source_col, x_col, y_col, z_col, color_col]
    missing = [col for col in required if col not in data.columns]
    if missing:
        raise KeyError(f"Missing selected column(s): {missing}")

    c_original = to_number(data[color_col])
    excluded_for_log = 0
    if use_log_color:
        excluded_for_log = int((~(c_original > 0)).sum())
        c_plot = pd.Series(np.where(c_original > 0, np.log(c_original), np.nan), index=data.index)
    else:
        c_plot = c_original

    file_name_series = (
        data[SYNTHETIC_FILE_COL]
        if SYNTHETIC_FILE_COL in data.columns
        else pd.Series("", index=data.index)
    )
    file_row_series = (
        data[SYNTHETIC_ROW_COL]
        if SYNTHETIC_ROW_COL in data.columns
        else pd.Series("", index=data.index)
    )

    cleaned = pd.DataFrame({
        "source": data[source_col].fillna("").astype(str).str.strip(),
        "file_name": file_name_series.fillna("").astype(str).str.strip(),
        "file_row": file_row_series.fillna("").astype(str).str.strip(),
        "x": to_number(data[x_col]),
        "y": to_number(data[y_col]),
        "z": to_number(data[z_col]),
        "c_original": c_original,
        "c_plot": c_plot,
    })

    finite_mask = (
        cleaned["source"].ne("")
        & np.isfinite(cleaned["x"])
        & np.isfinite(cleaned["y"])
        & np.isfinite(cleaned["z"])
        & np.isfinite(cleaned["c_original"])
        & np.isfinite(cleaned["c_plot"])
    )
    return cleaned.loc[finite_mask].copy(), excluded_for_log


def make_3d_scatter(
    data: pd.DataFrame,
    *,
    source_col: str,
    x_col: str,
    y_col: str,
    z_col: str,
    color_col: str,
    color_scale_name: str = "Cool",
    show_colorbar: bool = True,
    use_log_color: bool = False,
    selected_sources: Sequence[str] | None = None,
) -> tuple[go.Figure, str]:
    """Create the Plotly 3D scatter figure and a status message."""
    cleaned, excluded_for_log = prepare_plot_data(
        data,
        source_col=source_col,
        x_col=x_col,
        y_col=y_col,
        z_col=z_col,
        color_col=color_col,
        use_log_color=use_log_color,
    )

    if cleaned.empty:
        detail = " When ln(colour value) is enabled, colour values must be positive." if use_log_color else ""
        raise ValueError(f"No valid numeric rows found for the selected columns.{detail}")

    if selected_sources:
        selected_sources = list(selected_sources)
        cleaned = cleaned[cleaned["source"].isin(selected_sources)].copy()
        if cleaned.empty:
            raise ValueError("No rows remain after applying the selected source filter.")

    colour_label = f"ln({color_col})" if use_log_color else color_col
    colorscale = get_selected_color_scale(color_scale_name)
    sources = list(pd.unique(cleaned["source"]))
    cmin = float(cleaned["c_plot"].min())
    cmax = float(cleaned["c_plot"].max())

    fig = go.Figure()
    for index, source in enumerate(sources):
        part = cleaned[cleaned["source"] == source]
        customdata = np.column_stack([
            part["c_original"],
            part["c_plot"],
            part["file_name"],
            part["file_row"],
        ])
        fig.add_trace(
            go.Scatter3d(
                type="scatter3d",
                mode="markers",
                name=str(source),
                x=part["x"],
                y=part["y"],
                z=part["z"],
                customdata=customdata,
                marker=dict(
                    size=3.5,
                    color=part["c_plot"],
                    colorscale=colorscale,
                    cmin=cmin,
                    cmax=cmax,
                    opacity=0.82,
                    showscale=show_colorbar and index == 0,
                    colorbar=dict(
                        title=dict(text=colour_label, side="top"),
                        orientation="h",
                        x=0.5,
                        xanchor="center",
                        y=-0.25,
                        yanchor="top",
                        len=0.75,
                        thickness=20,
                    ),
                ),
                hovertemplate=(
                    "<b>%{fullData.name}</b><br>"
                    "File: %{customdata[2]}<br>"
                    "CSV row: %{customdata[3]}<br>"
                    f"{x_col}: %{{x}}<br>"
                    f"{y_col}: %{{y}}<br>"
                    f"{z_col}: %{{z}}<br>"
                    f"{color_col}: %{{customdata[0]}}<br>"
                    + (f"ln({color_col}): %{{customdata[1]}}<br>" if use_log_color else "")
                    + "<extra></extra>"
                ),
            )
        )

    fig.update_layout(
        title="Interactive 3D scatter plot",
        scene=dict(
            xaxis=dict(title=f"X: {x_col}"),
            yaxis=dict(title=f"Y: {y_col}"),
            zaxis=dict(title=f"Z: {z_col}"),
        ),
        margin=dict(l=0, r=0, t=55, b=150),
        legend=dict(title=dict(text="Source")),
        template="plotly_white",
        height=840,
    )

    log_message = ""
    if use_log_color and excluded_for_log > 0:
        log_message = f" {excluded_for_log:,} rows were excluded because the colour value was not positive."

    status = f"Rendered {len(cleaned):,} points across {len(sources):,} sources.{log_message}"
    return fig, status
````

## Optional JupyterLab widget interface

Run the next two cells to use upload buttons, column dropdowns, colour-map controls, and source filtering directly inside the notebook.

````python
def create_viewer_app(viewer: CHFCSVViewer | None = None):
    """Create an ipywidgets interface similar to the original HTML controls."""
    if widgets is None:
        raise ImportError("ipywidgets is not available. Install it with `%pip install ipywidgets`.")

    viewer = viewer or CHFCSVViewer()

    upload = widgets.FileUpload(accept=".csv,text/csv", multiple=True, description="Upload CSV")
    clear_btn = widgets.Button(description="Clear loaded data", button_style="danger", disabled=True)
    render_btn = widgets.Button(description="Render 3D plot", button_style="primary", disabled=True)

    source_dropdown = widgets.Dropdown(description="Source")
    x_dropdown = widgets.Dropdown(description="X")
    y_dropdown = widgets.Dropdown(description="Y")
    z_dropdown = widgets.Dropdown(description="Z")
    color_dropdown = widgets.Dropdown(description="Colour")

    log_toggle = widgets.Checkbox(value=False, description="Apply ln(colour value)")
    colorbar_toggle = widgets.Checkbox(value=True, description="Show colour bar")
    color_map_dropdown = widgets.Dropdown(options=list(COLOR_SCALES), value="Cool", description="Colour map")
    source_filter = widgets.SelectMultiple(options=[], description="Sources", rows=8)

    status = widgets.HTML("<em>No CSV loaded.</em>")
    mapping_summary = widgets.HTML("")
    file_summary_output = widgets.Output()
    plot_output = widgets.Output()

    def refresh_file_summary():
        with file_summary_output:
            clear_output()
            if viewer.loaded_files:
                display(viewer.file_summary())

    def refresh_controls():
        if not viewer.headers:
            render_btn.disabled = True
            clear_btn.disabled = True
            return

        options = viewer.headers
        mapping = viewer.default_mapping()
        for dropdown in (source_dropdown, x_dropdown, y_dropdown, z_dropdown, color_dropdown):
            dropdown.options = options

        source_dropdown.value = mapping.get("source", options[0])
        x_dropdown.value = mapping.get("x", options[0])
        y_dropdown.value = mapping.get("y", options[0])
        z_dropdown.value = mapping.get("z", options[0])
        color_dropdown.value = mapping.get("color", options[0])
        render_btn.disabled = False
        clear_btn.disabled = False
        status.value = f"Loaded {len(viewer.data):,} rows from {len(viewer.loaded_files):,} file(s)."
        refresh_file_summary()

    def refresh_sources_from_current_selection():
        if viewer.data.empty or not source_dropdown.value:
            source_filter.options = []
            source_filter.value = ()
            return
        sources = sorted(viewer.data[source_dropdown.value].fillna("").astype(str).str.strip().replace("", np.nan).dropna().unique())
        source_filter.options = sources
        source_filter.value = tuple(sources)

    def on_upload_change(change):
        value = change.get("new")
        if not value:
            return
        viewer.load_uploaded(value)
        refresh_controls()
        refresh_sources_from_current_selection()

    def on_clear_click(_):
        viewer.clear()
        for dropdown in (source_dropdown, x_dropdown, y_dropdown, z_dropdown, color_dropdown):
            dropdown.options = []
        source_filter.options = []
        source_filter.value = ()
        render_btn.disabled = True
        clear_btn.disabled = True
        status.value = "<em>No CSV loaded.</em>"
        mapping_summary.value = ""
        with file_summary_output:
            clear_output()
        with plot_output:
            clear_output()

    def on_render_click(_=None):
        if viewer.data.empty:
            status.value = "<strong>No CSV loaded.</strong>"
            return
        selected_sources = list(source_filter.value) or None
        colour_label = f"ln({color_dropdown.value})" if log_toggle.value else color_dropdown.value
        mapping_summary.value = (
            f"<strong>Current mapping:</strong> Source = {source_dropdown.value} &nbsp; | &nbsp; "
            f"X = {x_dropdown.value} &nbsp; | &nbsp; Y = {y_dropdown.value} &nbsp; | &nbsp; "
            f"Z = {z_dropdown.value} &nbsp; | &nbsp; Colour = {colour_label} &nbsp; | &nbsp; "
            f"Colour map = {color_map_dropdown.value}"
        )
        with plot_output:
            clear_output()
            try:
                fig, message = make_3d_scatter(
                    viewer.data,
                    source_col=source_dropdown.value,
                    x_col=x_dropdown.value,
                    y_col=y_dropdown.value,
                    z_col=z_dropdown.value,
                    color_col=color_dropdown.value,
                    color_scale_name=color_map_dropdown.value,
                    show_colorbar=colorbar_toggle.value,
                    use_log_color=log_toggle.value,
                    selected_sources=selected_sources,
                )
                status.value = message
                display(fig)
                viewer.has_rendered_plot = True
            except Exception as exc:
                status.value = f"<strong>Plot error:</strong> {exc}"

    upload.observe(on_upload_change, names="value")
    clear_btn.on_click(on_clear_click)
    render_btn.on_click(on_render_click)
    source_dropdown.observe(lambda change: refresh_sources_from_current_selection(), names="value")

    controls = widgets.VBox([
        widgets.HTML("<h3>Offline 3D CHF multi-CSV viewer</h3>"),
        upload,
        status,
        file_summary_output,
        widgets.HBox([clear_btn, render_btn]),
        widgets.GridBox(
            [source_dropdown, x_dropdown, y_dropdown, z_dropdown, color_dropdown],
            layout=widgets.Layout(grid_template_columns="repeat(2, minmax(260px, 1fr))", grid_gap="8px"),
        ),
        widgets.HBox([log_toggle, colorbar_toggle, color_map_dropdown]),
        widgets.HTML("<strong>Source filter</strong>"),
        source_filter,
        mapping_summary,
        plot_output,
    ])
    controls.viewer = viewer
    return controls
````

````python
# Run this cell in JupyterLab to launch the widget version.
app = create_viewer_app()
display(app)
````

## Notes on conversion

The original HTML contained a large embedded Plotly JavaScript bundle plus a smaller custom script. This notebook converts the custom script’s data handling and plotting behaviour to Python. Plotly itself is imported as a Python dependency rather than embedded in the notebook.
